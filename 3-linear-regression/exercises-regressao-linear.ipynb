```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from matplotlib.pyplot import subplots
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,summarize,poly)
```

```python
auto['horsepower'] = auto['horsepower'].replace('?', np.nan)
auto['horsepower'] = pd.to_numeric(auto['horsepower'])
auto = auto.dropna(subset=['horsepower'])
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

y = auto['mpg']
X = sm.add_constant(auto['horsepower'])


# Criar o modelo OLS
model1 = sm.OLS(y, X).fit() 

# Resumo dos resultados
print(model1.summary())
```

Temos um pvalor menor que 0.05 logo podemos falar que existe uma relao. Nosso r^2  de 60,6%, ou seja 60,2% da variabilidade de mpg pode ser explicada com horsrpower. Temos uma relao negativa, ou seja, a medida que o horsepower aumenta o mpg diminui.

```python
# Supondo que 'model1'  o modelo de regresso ajustado

# Criar um novo DataFrame com a constante (intercepto) e o valor de 'horsepower' de interesse (98)
new_data = pd.DataFrame({'const': 1, 'horsepower': 98}, index=[0])

# Prever 'mpg' para a potncia de 98
predicted_mpg = model1.predict(new_data)

# Obter os intervalos de confiana e previso
conf_int = model1.conf_int(alpha=0.05)
pred_int = model1.get_prediction(new_data).conf_int(alpha=0.05)

# Imprimir os resultados
print("mpg previsto:", predicted_mpg[0])
print("Intervalo de Confiana de 95% para mpg:", (conf_int.loc['horsepower'][0], conf_int.loc['horsepower'][1]))
print("Intervalo de Previso de 95% para mpg:", (pred_int[0][0], pred_int[0][1]))
```

results. (You can perform regression without an intercept using the keywords argument intercept=False to ModelSpec().)

```python
#regresso linear sem intercepto
model = sm.OLS(y, x, hasconst=False).fit()

print(model.summary())

# Coeficiente estimado ()
coef_estimate = model.params[0]

# Erro padro do coeficiente estimado
std_error = model.bse[0]

# T-statistic
t_statistic = coef_estimate / std_error

# P-value associado  hiptese nula H0:  = 0
p_value = model.pvalues[0]

print("Coeficiente estimado:", coef_estimate)
print("Erro padro:", std_error)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

Isso ocorrer quando a relao entre X e Y for uma relao perfeitamente linear, ou seja, quando os pontos (X, Y) estiverem alinhados em uma linha reta. Nesse caso, no importa se estamos prevendo Y com base em X ou X com base em Y, porque a relao  a mesma em ambas as direes.

b) Generate an example in Python with n = 100 observations in which the coefcient estimate for the regression of X onto Y
is diferent from the coefcient estimate for the regression of Y onto X.

```python
# Gerando os dados
np.random.seed(2)
n = 100
X = np.random.rand(n) * 10
Y = 2 * X + np.random.randn(n) * 3

# Ajustar os modelos 
model_X_on_Y = sm.OLS(X, sm.add_constant(Y)).fit()
model_Y_on_X = sm.OLS(Y, sm.add_constant(X)).fit()

# Coeficientes estimados
coef_X_on_Y = model_X_on_Y.params
coef_Y_on_X = model_Y_on_X.params

print("Coeficientes para a regresso de X sobre Y:", coef_X_on_Y)
print("Coeficientes para a regresso de Y sobre X:", coef_Y_on_X)

#Grficos
plt.scatter(Y, X, label='Dados')
plt.plot(Y, model_X_on_Y.predict(sm.add_constant(Y)), color='red', label='X sobre Y')
plt.plot(X, model_Y_on_X.predict(sm.add_constant(X)), color='green', label='Y sobre X')
plt.xlabel('Y')
plt.ylabel('X')
plt.legend()
plt.show()
```
