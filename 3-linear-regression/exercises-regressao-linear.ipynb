```python
auto['horsepower'] = auto['horsepower'].replace('?', np.nan)
auto['horsepower'] = pd.to_numeric(auto['horsepower'])
auto = auto.dropna(subset=['horsepower'])
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

y = auto['mpg']
X = sm.add_constant(auto['horsepower'])


# Criar o modelo OLS
model1 = sm.OLS(y, X).fit() 

# Resumo dos resultados
print(model1.summary())
```

Temos um pvalor menor que 0.05 logo podemos falar que existe uma relação. Nosso r^2 é de 60,6%, ou seja 60,2% da variabilidade de mpg pode ser explicada com horsrpower. Temos uma relação negativa, ou seja, a medida que o horsepower aumenta o mpg diminui.

```python
# Supondo que 'model1' é o modelo de regressão ajustado

# Criar um novo DataFrame com a constante (intercepto) e o valor de 'horsepower' de interesse (98)
new_data = pd.DataFrame({'const': 1, 'horsepower': 98}, index=[0])

# Prever 'mpg' para a potência de 98
predicted_mpg = model1.predict(new_data)

# Obter os intervalos de confiança e previsão
conf_int = model1.conf_int(alpha=0.05)
pred_int = model1.get_prediction(new_data).conf_int(alpha=0.05)

# Imprimir os resultados
print("mpg previsto:", predicted_mpg[0])
print("Intervalo de Confiança de 95% para mpg:", (conf_int.loc['horsepower'][0], conf_int.loc['horsepower'][1]))
print("Intervalo de Previsão de 95% para mpg:", (pred_int[0][0], pred_int[0][1]))
```

(a) Perform a simple linear regression of y onto x, without an intercept. Report the coefcient estimate βˆ, the standard error of this coefcient estimate, and the t-statistic and p-value associated with the null hypothesis H0 : β = 0. Comment on these
results. (You can perform regression without an intercept using the keywords argument intercept=False to ModelSpec().)

```python
#regressão linear sem intercepto
model = sm.OLS(y, x, hasconst=False).fit()

print(model.summary())

# Coeficiente estimado (βˆ)
coef_estimate = model.params[0]

# Erro padrão do coeficiente estimado
std_error = model.bse[0]

# T-statistic
t_statistic = coef_estimate / std_error

# P-value associado à hipótese nula H0: β = 0
p_value = model.pvalues[0]

print("Coeficiente estimado:", coef_estimate)
print("Erro padrão:", std_error)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
