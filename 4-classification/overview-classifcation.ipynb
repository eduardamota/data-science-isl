A odds em regressão logística são definidas como a proporção da probabilidade de sucesso (evento ocorrendo) em relação à probabilidade de fracasso (evento não ocorrendo).

Os odds podem assumir qualquer valor entre 0 e ∞. Aqui estão algumas interpretações importantes dos odds:

Quando os odds são próximos de 0, isso indica que a probabilidade do evento ocorrer é muito baixa em relação à probabilidade de não ocorrer. Isso significa que o evento é considerado improvável.
Quando os odds são grandes (próximos a ∞), isso indica que a probabilidade do evento ocorrer é muito alta em relação à probabilidade de não ocorrer. Isso significa que o evento é considerado altamente provável.

### Estimating the Regression Coefcients

Em regressão linear usamos a abordagem dos mínimos quadrados para estimar os coeficientes desconhecidos. Embora possamos usar (non-linear) minimos quadrados para ajustar um modelo, porém existem modelos mais generalistas como maximum likelihood.

**maximum likelihood** a lógica por trás desse método é que vamos procurar por a estimativa β0 e β1que tornem as previsões de probabilidade mais próximas das individuais.

### Making Predictions

Uma vez que os coeficients foram estimados, podemos calcular a probabilidade.

Podemos usar preditores qualitativos na regressão logistica usando dummy. Um exemplo que está no dataset do livro contém a variável estudante. Para ajustar um modelo que usa o status do estudante como variável preditora, podemos criar uma var dummy que aplica o valor 1 para estudantes e 0 ara não estudantes.

### Multiple Logistic Regression

Sabemos que ainda dentro da regressão linear os resultados usando um variável preditora podem ser um tanto quanto diferente quando são obtidos usando multiplos preditores, especialmente quando há uma correlação entre esses preditores. Em geral, esse fenomeno dentro da regressão logística é chamado de confounding.
