np.ix_() é usada para criar um "objeto de indexação avançada", que é uma tupla de índices que será usada para realizar uma indexação específica em uma matriz (array multidimensional) do NumPy

### Boolean Indexing

```python
 keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
 keep_rows[[1,3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([0,1,0,1])) #Função all checa se todas as entradas são verdadeiras
```

```python
 A[np.array([0,1,0,1])]
```

O array de índices [0, 1, 0, 1] indica que queremos selecionar as linhas 0, 1, 0 novamente (repetição) e 1 novamente (repetição).

```python
 A[keep_rows]
```

```python
#Criando um arrau de zeros com o número de colunas iguala matriz A
#O A.shpae retorna uma tupla informando o número de linhas e colunas, logo A.shape[1] extrai o número de colunas
keep_cols = np.zeros(A.shape[1], bool) 

#Vamos pegar as colunas 0,2,3 e adicionar TRUE
keep_cols[[0, 2, 3]] = True

# np.ix_(), podemos criar um índice avançado (advanced index) 
#para selecionar as células específicas da matriz com base nas listas booleanas
idx_bool = np.ix_(keep_rows, keep_cols)

A[idx_bool]
```

```python
#Podemos também fazer um mix de uma lista e um array de boleanos usando np.ix()
idx_mixed = np.ix_([1,3], keep_cols)
A[idx_mixed]
```

### More on Selecting Rows and Columns

Vamos supor que eu queira criar um dataframe com weight e na origin do subconjunto de carros com ano maior que 80 - ou seja, aqueles construídos após 1980.

```python
idx_80 = auto_re['year']>80
auto_re.loc[idx_80,['weight','origin']].head()
```

```python
#Podemos fazer isso usando lambda
auto_re.loc[lambda x: x['year']>80,['weight','origin']].head()
```

Vamos supor que eu queira criar um dataframe com weight e na origin do subconjunto de carros com ano maior que 80 - ou seja, aqueles construídos após 1980 e o mpg maior que 30.

```python
auto_re.loc[lambda x: (x['year']>80) & (x['mpg']>30),['weight','origin']].head()
```

Agora queremos todos os Fords e Datsun com displacement menores que 300.

```python
#Podemos usar str.contains()

auto_re.loc[lambda df: (df['displacement']<300)
                        & (df.index.str.contains('ford'))
                        | (df.index.str.contains('datsun')),
                        ['weight','origin']].head()
```

## For Loops 

```python
total = 0
for value in [3,2,19]:
    total += value
print('Total é: {0}'.format(total))
```

```python
total = 0
for value in [2,3,19]:
    for weight in [3, 2, 1]:
        total += value * weight
print('Total is: {0}'.format(total))
```

```python
total = 0
for value, weight in zip ([2,3,19],[0.2,0.3,0.5]):
    total+=value*weight
print('Weighted average is: {0}'.format(total))
```

###   String Formatting

```python
#Cria número aleatorios
rng = np.random.default_rng(1)

#matriz A com dimensões (127, 5) preenchida com números aleatórios de uma distribuição normal padrão
A = rng.standard_normal((127, 5)) 


#Lista aleatoria [0,p.nan], onde a probabilidade de selecionar 0 é de 80% e np.nan é de 20%.
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape)

A+= M

D = pd.DataFrame(A, columns=['food',
'bar',
'pickle',
'snack',
'popcorn'])


D[:3]
```

```python
#Vamos criar um loop para imprimir a porcentagem que falta em cada coluna

for col in D.columns:
    template = 'Column "{0}" has {1:.2%} missing values' #0 é nome da coluna e 2 porcentagem de valores ausentes
    print(template.format(col,
        np.isnan(D[col]).mean()))
```
